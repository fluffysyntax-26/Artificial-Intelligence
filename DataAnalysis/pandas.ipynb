```python
import pandas as pd
import numpy as np
```

```python
# Series - One dimensional Arrary, usually the column

# convert a list into a pandas series
data = [1,2,3,4,5]
series = pd.Series(data)
series
```

```python
# Create a series from dictionary

user_profile = {
    "username": "john_doe",
    "is_active": True,
    "login_count": 45,
    "last_login": "2025-08-09",
    "favorite_numbers": [7, 13, 21],
    "contact": {
        "email": "john.doe@example.com",
        "phone": "555-1234"
    }
}

user = pd.Series(user_profile)
user
```

```python
data = [10, 20, 30]
index = ['a', 'b', 'c']

pd.Series(data, index=index)
```

```python
# DataFrame - Two Dimensional Array, similar to a table in a database

# Create a dataframe from a dictionary fo lists

student_subjects = {
    "John": ["Mathematics", "Physics", "Chemistry"],
    "Alice": ["History", "Literature", "Calculus"],
    "Bob": ["Computer Science", "Physics", "Chemistry"]
}

df = pd.DataFrame(student_subjects)
df
```

```python
# you can convert a dataframe into an array using numpy

np.array(df)
```

```python
# Data mamipulation with DataFrame
student_roster = [
    {
        "id": "101",
        "name": "Alice Johnson",
        "major": "Computer Science"
    },
    {
        "id": "102",
        "name": "Bob Williams",
        "major": "Mechanical Engineering"
    },
    {
        "id": "103",
        "name": "Charlie Brown",
        "major": "Business Administration"
    }
]


df = pd.DataFrame(student_roster)
df
```

```python
# Add columns to the dataframe

df['City'] = ['New York', 'Boston', "LA"]
df["Age"] = [20, 19, 21]
df
```

```python
# Drop column, axis = 1; axis = 0 for rows
# to save it, inplace = True
df.drop('City', axis=1, inplace=True)
df
```

```python
# Add age to the column
df["Age"] = df["Age"] + 1
df
```
